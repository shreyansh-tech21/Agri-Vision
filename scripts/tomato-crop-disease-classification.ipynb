```python
!pip install ultralytics -q
from ultralytics import YOLO
data_path = "refer-to-readme-for-dataset-info"

model = YOLO("yolov8n.pt")

model.train(
    data=data_path,
    epochs=25,
    imgsz=640,
    batch=16,
    device="0,1"
)


metrics = model.val()

results = model.predict(
    source="refer-to-readme-for-dataset-info",
    save=True,
    conf=0.25
)
model.export(format="onnx")
```
